# **Popularity Based recommendation system**

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

```python
print(ratings_with_name.dtypes)
```

```python
ratings_with_name['Book-Rating'].apply(np.int64)
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
num_rating_df[num_rating_df['num_ratings'] >=250]
```

```python
num_rating_df = num_rating_df.merge(books , on='Book-Title').drop_duplicates('Book-Title')[['Book-Title' ,'Book-Author' , 'Image-URL-M', 'num_ratings' ]]
```

# **Collaborative filtering Based recommendation system**

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
educated_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(educated_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0 ,inplace=True )
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
sorted(list(enumerate(similarity_score[0])), key = lambda x:x[1] , reverse=True [1:6])
```

```python
def recommened(book_name):
    #index fetch
    index = np.where(pt.index== book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1] , reverse=True) [1:6]
    data = []
    for i in similar_items:
      item = []
      temp_df = books[books['Book-Title'] == pt.index[i[0]]]
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'] ))
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'] ))
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M']))
      data.append(item)
    return data
```
